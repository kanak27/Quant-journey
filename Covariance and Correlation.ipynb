```python
import pandas as pd
import yfinance as yfin
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
import seaborn as sns
```

```python
tickers = ['RELIANCE.NS', 'HDFCBANK.NS', 'TCS.NS', 'INFY.NS', 'ITC.NS']

start = dt.datetime.now().replace(year=dt.datetime.now().year - 1)
end = dt.datetime.now()

nifty5 = yfin.download(tickers, start = start, end = end)
nifty5
```

```python
nifty5[('DailyReturns', 'REL')] = nifty5[('Close', 'RELIANCE.NS')].pct_change(1) * 100
nifty5[('DailyReturns', 'HDFC')] = nifty5[('Close', 'HDFCBANK.NS')].pct_change(1) * 100
nifty5[('DailyReturns', 'INF')] = nifty5[('Close', 'INFY.NS')].pct_change(1) * 100
nifty5[('DailyReturns', 'ITC')] = nifty5[('Close', 'ITC.NS')].pct_change(1) * 100
nifty5[('DailyReturns', 'TCS')] = nifty5[('Close', 'TCS.NS')].pct_change(1) * 100
```

```python
nifty5.replace(np.nan, 0, inplace = True)
nifty5
```

```python
dailyReturns = nifty5.DailyReturns
dailyReturns = dailyReturns.reset_index()
dailyReturns = dailyReturns.drop('Date', axis=1)   
dailyReturns
```

```python
covarianceMatrix = dailyReturns.cov()
```

```python
correlationMatrix = dailyReturns.corr()
```

```python
sns.heatmap(covarianceMatrix, annot = True)
```

```python
sns.heatmap(correlationMatrix, annot = True)
```
